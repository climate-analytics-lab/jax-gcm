```python
import jax
import jax.numpy as jnp

def my_func(x):  
    clipped = jnp.clip(x, -0.999, 0.999)
    return jnp.arccos(clipped)

# Define a function to calculate the gradient
def grad_my_func(f, x): # Parameters should be the same as my_func
    primals, f_vjp = jax.vjp(f, x)  # Creates primals and "derivative function"
    input = (jnp.ones_like(primals)) # Input into f_vjp.  Must be the shape of the output of my_func 
                                # I think it should be initialized with all ones to get the gradient value with
                                # respect to the parameters of my_func 
    df_dx = f_vjp(input) #Takes derivate with respect to each parameter
    return df_dx
```

```python
#Test
xx = jnp.array([-2.0, -1.0, 0.0, 1.0, 2.0])
print(my_func(xx))
df_dx = grad_my_func(my_func, xx)

print(df_dx)
```
