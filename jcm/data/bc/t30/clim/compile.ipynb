```python
file = ["land.nc", "sea_ice.nc", "sea_surface_temperature.nc", "snow.nc", "soil.nc", "surface.nc"]
names = ["land", "sea_ice", "sea_surface", "snow", "soil", "surface"]
    
twod = ["orog", "lsm", "vegh", "alb", "vegl"]
datasets={}
    # Load each dataset
for file_path, name in zip(file, names):
    try:
        # Open the dataset
        ds = xr.open_dataset(file_path)
        
        # Store in dictionary with prefix to avoid naming conflicts
        for var in ds.data_vars:
            # if var in twod:
            #     datasets[f"{var}"] = ds[var].transpose('lon', 'lat')
            # else:
            datasets[f"{var}"] = ds[var].transpose()
            
        # Close the dataset to free memory
        ds.close()
        
    except Exception as e:
        print(f"Error loading {file_path}: {str(e)}")
        continue

# Combine all variables into a single dataset
```

```python
combined_ds = xr.Dataset(datasets)
```

```python
combined_ds.to_netcdf("boundaries.nc")
```
